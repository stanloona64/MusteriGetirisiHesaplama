```python
result=df.groupby(["COUNTRY", "SOURCE", "SEX", "AGE"]).agg({"PRICE": "mean"})
result["PRICE"] = result["PRICE"].round(2)
agg_df = result.sort_values(by="PRICE", ascending=False)
agg_df = agg_df.reset_index()
agg_df
```

```python
bins = [0, 18, 23, 30, 40, 70]
labels = ["0_18", "19_23", "24_30", "31_40", "41_70"]
agg_df["AGE_CAT"] = pd.cut(agg_df["AGE"], bins=bins, labels=labels, right=False)
agg_df
```

```python
agg_df['customers_level_based'] = (agg_df["COUNTRY"].str.upper() + '_' +
                                   agg_df["SOURCE"].str.upper() + '_' +
                                   agg_df["SEX"].str.upper() + '_' +
                                   agg_df["AGE_CAT"].str.upper())
result = agg_df.groupby("customers_level_based")["PRICE"].mean().reset_index()
result["PRICE"] = result["PRICE"].round(2)
result = result.sort_values(by="PRICE", ascending=False)
result
```

```python
labels = ["D", "C", "B", "A"]
agg_df["AGE_CAT"] = pd.qcut(agg_df["AGE"],4,labels=labels)
agg_df
```
